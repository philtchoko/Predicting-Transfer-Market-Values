```python
import streamlit as st
import joblib
import numpy as np
import pandas as pd

pipeline = joblib.load("final_transfer_value_model.pkl")

st.title("⚽ Transfer Market Value Predictor")

st.header("Enter Player Stats:")

# User Inputs
goals_assists = st.number_input("Goals + Assists (G+A)", min_value=0, max_value=100, value=10)
expected_contribution = st.number_input("Expected Contribution (xG + xAG)", min_value=0.0, max_value=50.0, value=5.0)
shots = st.number_input("Total Shots (Sh_Standard)", min_value=0, max_value=200, value=30)
position = st.selectbox("Position", ["FW", "MF", "DF"])
age_group = st.selectbox("Age Group", ["Young", "Prime", "Veteran"])

# Create input DataFrame
input_data = pd.DataFrame({
    "G+A": [goals_assists],
    "Expected_Contribution": [expected_contribution],
    "Sh_Standard": [shots],
    "Position": [position],
    "Age_Group": [age_group]
})

# Predict button
if st.button("Predict Transfer Value (€)"):
    prediction = pipeline.predict(input_data)[0]
    st.success(f"Estimated Market Value: €{prediction:,.0f}")
```

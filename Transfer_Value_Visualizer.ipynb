```python
import gradio as gr
import pandas as pd
import joblib

# Load pipeline
pipeline = joblib.load('/Users/philtchoko/Desktop/DATS4001/final_transfer_value_model.pkl')  # Adjusted for Colab

def predict_transfer_value(position, age, ga=0, expected_contribution=0.0, shots=0,
                           assists=0, xag=0.0, kp=0, prgp=0, prgc=0,
                           tackles=0, blocks=0, pass_blocks=0):
    if position == "FW":
        data = pd.DataFrame([{
            "G+A": ga,
            "Expected_Contribution": expected_contribution,
            "Sh_Standard": shots,
            "Position": position,
            "Age_Group": age
        }])
    elif position == "MF":
        data = pd.DataFrame([{
            "Ast": assists,
            "xAG": xag,
            "KP": kp,
            "PrgP": prgp,
            "PrgC_Carries": prgc,
            "Position": position,
            "Age_Group": age
        }])
    elif position == "DF":
        data = pd.DataFrame([{
            "Att 3rd_Tackles": tackles,
            "Blocks_Blocks": blocks,
            "Pass_Blocks": pass_blocks,
            "Position": position,
            "Age_Group": age
        }])
    else:
        return "Unknown position"

    prediction = pipeline.predict(data)[0]
    return f"Estimated Market Value: €{prediction:,.0f}"

position_input = gr.Dropdown(["Forward", "Midfielder", "Defender"], label="Position")
age_group_input = gr.Dropdown(["Young", "Prime", "Veteran"], label="Age Group")

with gr.Blocks() as demo:
    gr.Markdown("### ⚽ Soccer Player Transfer Value Estimator")

    with gr.Row():
        position = position_input.render()
        age = age_group_input.render()

    with gr.Row():
        ga = gr.Number(label="Goals + Assists", value=0)
        expected_contribution = gr.Number(label="Expected Contribution (xG + xAG)", value=0.0)
        shots = gr.Number(label="Shots", value=0)

    with gr.Row():
        assists = gr.Number(label="Assists", value=0)
        xag = gr.Number(label="xAG", value=0.0)
        kp = gr.Number(label="Key Passes (KP)", value=0)
        prgp = gr.Number(label="Progressive Passes (PrgP)", value=0)
        prgc = gr.Number(label="Progressive Carries (PrgC_Carries)", value=0)

    with gr.Row():
        tackles = gr.Number(label="Tackles in Final Third", value=0)
        blocks = gr.Number(label="Blocks", value=0)
        pass_blocks = gr.Number(label="Pass Blocks", value=0)

    predict_button = gr.Button("Predict Value")
    output = gr.Textbox(label="Estimated Market Value")

    predict_button.click(predict_transfer_value,
                         [position, age, ga, expected_contribution, shots,
                          assists, xag, kp, prgp, prgc,
                          tackles, blocks, pass_blocks],
                         output)

demo.launch(share=True)
```
